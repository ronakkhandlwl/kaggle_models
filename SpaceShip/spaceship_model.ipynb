```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
def engineer_data(df, for_test=False):
    # df = df.dropna()
    df['HomePlanet'], _ = pd.factorize(df['HomePlanet'])
    df['CryoSleep'], _ = pd.factorize(df['CryoSleep'])
    df['VIP'], _ = pd.factorize(df['VIP'])
    df['Destination'], _ = pd.factorize(df['Destination'])
    df = df.drop(columns=['PassengerId', 'Cabin', 'Name'])
    if not for_test:
        df['Transported'], _ = pd.factorize(df['Transported'])
        X = df.drop(columns=['Transported'])
        y = df['Transported']
        return X, y
    return df
```

```python
X, y = engineer_data(df)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
model.fit(
      X_train, y_train,
      epochs=15,
      validation_data = (X_test, y_test),
      verbose=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier(n_estimators=100, random_state=42)
rfc_model.fit(X_train, y_train)
rfc_model.score(X_test, y_test)
```

```python
otdf = pd.read_csv('test.csv')
otdf.head()
```

```python
tdf = engineer_data(otdf, for_test=True)
tdf.head()
```

```python
test_pred = model.predict(tdf).ravel()
new_df = pd.DataFrame({'PassengerId': otdf['PassengerId'], 'Transported': test_pred>0.5})
new_df.to_csv('predictions2.csv', index=False)
```

```python
tp = rfc_model.predict(tdf)
new_df = pd.DataFrame({'PassengerId': otdf['PassengerId'], 'Transported': tp==1})
new_df.to_csv('predictions.csv', index=False)
```
