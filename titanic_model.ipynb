```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('titanic/train.csv')
```

```python
df.head(10)
```

```python
df['Embarked'], unique_values = pd.factorize(df['Embarked'])
df['Cabin'], unique_values_2 = pd.factorize(df['Cabin'])
df['Sex'], unique_values_3 = pd.factorize(df['Sex'])
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['Name', 'Ticket'])
df.shape
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=13)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
test_df = pd.read_csv('titanic/test.csv')
print(test_df.shape)
test_df['Embarked'], unique_values = pd.factorize(test_df['Embarked'])
test_df['Cabin'], unique_values_2 = pd.factorize(test_df['Cabin'])
test_df['Sex'], unique_values_3 = pd.factorize(test_df['Sex'])
test_df = test_df.drop(columns=['Name', 'Ticket'])
```

```python
test_df.head()
```

```python
test_pred = model.predict(test_df)
```
